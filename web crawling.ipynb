```python
import requests
from bs4 import BeautifulSoup
import re
import json
```

```python
def get_all_urls():
    urls = []
    for page in range(1,3):
        url = f"https://books.toscrape.com/catalogue/page-{page}.html"
        urls.append(url)
    return urls
    
def extract_product_links(urls):
    links = []
    for url in urls:
        res = requests.get(url)
        soup = BeautifulSoup(res.content, 'lxml')
        books = soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
        for book in books:
            link = "https://books.toscrape.com/catalogue/"+book.article.div.a['href']
            links.append(link)
    return links 
  
def get_details(books_links):
    detail = []
    all_books_details = []
    Dict= {"Name": " ",
          "UPC": " ",
          "Product Type": " ",
          "Price (excl. tax)": " ",
          "Price (incl. tax)": " ",
          "Tax": " ",
          "Availability": " ",
          "Number of reviews": " "}
    
    #opening jason file
    jsonFile = open("data.json", "w")
    
    #getting name and other details of each book
    for book_link in books_links:
        res = requests.get(book_link)
        soup = BeautifulSoup(res.content, 'lxml')
        book_name = soup.find('h1').text
        book_details = soup.find_all('tr')
        for i in range(0,7):
            detail.append(book_details[i].find('td').text)
        Dict= {"Name": book_name,
               "UPC": detail[0],
               "Product Type": detail[1],
               "Price (excl. tax)": detail[2],
               "Price (incl. tax)": detail[3],
               "Tax": detail[4],
               "Availability": detail[5],
               "Number of reviews": detail[6]}
        #appending all data in 1 array
        all_books_details.append(Dict)
        detail.clear()
    
    #writing data to json file
    jsonString = json.dumps(all_books_details)
    jsonFile = open("data.json", "w")
    jsonFile.write(jsonString)    
    jsonFile.close()
  
            #details.append(detail)
        #all_books_details.append(details)
        #all_books_details.append(details)
    #print(details)
        
    
            

#main function
if __name__ == '__main__':
    urls = get_all_urls()
    books_links = extract_product_links(urls)
    get_details(books_links)
    
```
